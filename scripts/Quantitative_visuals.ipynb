```python
# STEP 1: Install required libraries (only once in Colab)

# STEP 2: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files

# STEP 3: Upload your CSV file
uploaded = files.upload()

# STEP 4: Read the uploaded CSV
df = pd.read_csv(next(iter(uploaded)))

# STEP 5: Set visual style
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 5))

# STEP 6: Create bar chart with correct phase order
bar = sns.barplot(
    data=df,
    x='Question',
    y='Mean',
    hue='Phase',
    hue_order=['Before Refinement', 'After Refinement']  # âœ… Correct order
    palette='Set2',
    edgecolor='black'
)

# STEP 7: Add value labels on bars
for container in bar.containers:
    bar.bar_label(container, fmt="%.2f", label_type="edge", padding=2, fontsize=10)

# STEP 8: Customize appearance
plt.ylabel("Mean Likert Score")
plt.xlabel("")
plt.title("Requirement Quality Before and After Refinement", fontsize=13)
plt.xticks(rotation=30, ha='right')
plt.ylim(0, 5)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# STEP 9: Move legend outside the plot
plt.legend(title="Phase", loc='upper left', bbox_to_anchor=(1.02, 1), borderaxespad=0)

# STEP 10: Display chart
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Data from Question 11
data = {
    "Aspect": [
        "Correctness",
        "Clarity & wording",
        "Completeness",
        "Unnecessary requirements",
        "Missing requirements",
        "Business feasibility",
        "Technical feasibility"
    ],
    "Count": [0, 1, 4, 4, 5, 3, 4]
}

df_q11 = pd.DataFrame(data)

# Set academic style
sns.set_theme(style="whitegrid")

# Use a single academic blue color
bar_color = sns.color_palette("deep")[0]

# Create the plot
plt.figure(figsize=(10, 5))
bar = sns.barplot(
    data=df_q11.sort_values("Count", ascending=False),
    x="Aspect",
    y="Count",
    color=bar_color,
    edgecolor='black',
    width=0.6  # narrower bars
)

# Add count labels
for i, bar_patch in enumerate(bar.patches):
    height = bar_patch.get_height()
    bar.text(
        bar_patch.get_x() + bar_patch.get_width() / 2,
        height + 0.15 if height < 8 else height - 0.4,
        f"{int(height)}",
        ha='center',
        va='bottom' if height < 8 else 'top',
        color='black',
        fontsize=10
    )

# Final formatting
plt.title("", fontsize=13)
plt.ylabel("Number of Participants")
plt.xlabel("")
plt.ylim(0, 9)  # Ensure Y-axis represents total participants
plt.xticks(rotation=30, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
