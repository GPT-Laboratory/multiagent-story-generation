```python
# STEP 2: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# STEP 3: Upload your Excel file
from google.colab import files
uploaded = files.upload()

# STEP 4: Load the correct sheet and check available sheet names
xls = pd.ExcelFile(next(iter(uploaded)))
print("Available sheets:", xls.sheet_names)

# Load the Closed_ended sheet
df = pd.read_excel(xls, sheet_name='Closed_ended')

# STEP 5: Identify Likert-scale columns (adjust if needed)
likert_cols = [col for col in df.columns if "perceive the following statements" in col or "How likely are you" in col]

# Optional: Convert Likert values from text to numeric if needed
likert_map = {
    'Strongly Disagree (1)': 1,
    'Disagree (2)': 2,
    'Neutral (3)': 3,
    'Agree (4)': 4,
    'Strongly Agree (5)': 5
}

# Convert Likert-style strings to numbers
for col in likert_cols:
    if df[col].dtype == 'object':
        df[col] = df[col].map(likert_map)

# STEP 6: Summary statistics
summary = df[likert_cols].agg(['mean', 'std', 'count']).T
print("ðŸ“Š Summary Statistics:")
display(summary)

# STEP 7: Bar charts for each Likert-scale question
for col in likert_cols:
    plt.figure(figsize=(6, 3))
    sns.countplot(x=col, data=df, order=[1, 2, 3, 4, 5], palette="viridis")
    plt.title(col[:80] + "..." if len(col) > 80 else col)
    plt.xlabel("Likert Scale (1=Strongly Disagree to 5=Strongly Agree)")
    plt.ylabel("Count")
    plt.tight_layout()
    plt.show()
```
