```python
import nltk
from nltk.tokenize import sent_tokenize
from sentence_transformers import SentenceTransformer, util

# Initialize the model
model = SentenceTransformer('all-MiniLM-L6-v2')

# General project description
project_description = """
The Future Electrified Mobile Machines (FEMMa) project aims to develop electrified and autonomous mobile working machines to support the transition towards a sustainable and carbon-neutral future. The project explores various methodologies in data collection, analysis, and application, aiming to integrate these findings into industry practices effectively.
"""

# Tokenize the project description into sentences
nltk.download('punkt')
sentences = sent_tokenize(project_description)

# Define user stories for each model

# GPT 3.5 Model
user_stories_gpt = [
    "As a user, I want to access the main objectives of the FEMMa research project, to understand its primary aims.",
    "As a user, I want to know how the research data for the FEMMa project was selected and collected, to understand the data gathering process.",
    "As a user, I want to learn about the analysis methods used for data collection and analysis in the FEMMa project, to understand the research approach.",
    "As a user, I want to explore the project timeline of the FEMMa research project, to understand the key phases and timeline.",
    "As a user, I want to discover the key findings of the FEMMa research project, to understand the outcomes.",
    "As a user, I want to know the practical applications of the research findings from the FEMMa project, to understand their real-world implications.",
    "As a user, I want to understand how the research findings of the FEMMa project can impact industry practices or policies.",
    "As a user, I want to explore how the results of the FEMMa project can be utilized in the future.",
    "As a user, I want to learn about the societal impacts of the FEMMa project, to understand its broader implications.",
    "As a user, I want to access the list of publications related to the FEMMa research project.",
    "As a user, I want to know who funded the FEMMa research project."
]

# Omni Model
user_stories_omni = [
    "As a user, I want to view the main objectives of the FEMMa research project, to understand its primary goals.",
    "As a user, I want to know how the research data was selected and collected, to understand the methodology behind the data gathering process.",
    "As a user, I want to learn about the analysis methods used in the research, to comprehend how the data was analyzed.",
    "As a user, I want to view the project timeline and key phases, to get an overview of the project’s progression.",
    "As a user, I want to understand the key findings of the research, to grasp the main outcomes.",
    "As a user, I want to know the practical applications of the research findings, to see how they can be applied in real-world scenarios.",
    "As a user, I want to learn about the impact of the research findings on industry practices or policies, to understand its industrial relevance.",
    "As a user, I want to know how the project results can be utilized in the future, to see the potential for ongoing benefits.",
    "As a user, I want to understand the societal impacts of the FEMMa project, to appreciate its broader significance.",
    "As a user, I want to view the publications related to the FEMMa project, to access the detailed research work.",
    "As a user, I want to know who funded the FEMMa project, to understand the sources of financial support.",
    "As a user, I want to have a UI similar to ChatGPT with FEMMa’s logo, to have a familiar and branded interaction experience.",
    "As a user, I want to pre-select prompts provided on the UI, to quickly access common queries.",
    "As a user, I want to input a new prompt in the text area, to ask custom questions.",
    "As a developer, I want to export all the website’s data into multiple PDF files, to embed the data into the vector database.",
    "As a developer, I want to embed the data source into the vector database, to enable accurate and relevant responses.",
    "As a developer, I want to create an assistant that formats prompts with instructions and context, to ensure accurate and relevant responses.",
    "As a developer, I want to deploy the MVP on the University’s server, to make the application accessible."
]

# Llama3 70 billion Model
user_stories_llama = [

  "As a customer, I want to interact with a customized AI assistant that responds accurately and relevantly to my queries about the FEMMa project, so that I can get precise information about the project.",
  "As a user, I want to pre-select prompts provided on the UI or input a new prompt in the text area, so that I can easily initiate a new conversation with the AI assistant.",
  "As a developer, I want to export all the website's data into multiple pdf files, so that I can collect and process the data for the AI assistant.",
  "As a developer, I want to embed the collected data into a vector database, so that I can handle additions, editions, and removals of single/multiple pdf files from the database.",
  "As a developer, I want to create an assistant that exploits best practices to format prompts with the assistant's instructions and context, so that I can get the most accurate response from the AI assistant.",
  "As a developer, I want to develop the user interface using Streamlit, so that I can create a UI similar to ChatGPT with 'FEMMa's log'.",
  "As a developer, I want to deploy the MVP on the University's server, so that I can make the application available for users.",
  "As a user, I want to get accurate responses to my queries about the main objectives of the FEMMa research project, so that I can understand the project's goals.",
  "As a user, I want to get accurate responses to my queries about the data collection and analysis methods used in the FEMMa research project, so that I can understand the project's methodology.",
  "As a user, I want to get accurate responses to my queries about the project timeline, key findings, and practical applications of the FEMMa research project, so that I can understand the project's outcomes.",
  "As a user, I want to get accurate responses to my queries about the impact of the FEMMa research project on industry practices, policies, and society, so that I can understand the project's implications.",
  "As a user, I want to get accurate responses to my queries about the publications and funding of the FEMMa research project, so that I can understand the project's outputs and supporters."

]

# Mixtral Model
user_stories_mixtral = [
  "As a user, I want to view the main objectives of the research project, in order to understand the goals of the FEMMa project.",
  "As a user, I want to view the data collection methods, in order to understand how the research data was gathered and selected.",
  "As a user, I want to view the key findings of the research, in order to understand the results and outcomes of the project.",
  "As a user, I want to view the practical applications of the research, in order to understand how the findings can be applied in real-world situations.",
  "As a user, I want to view the publications of the project, in order to access the research findings and methodologies.",
  "As a user, I want to view the funding sources of the project, in order to understand the financial backing of the research.",
  "As a user, I want to input a custom prompt, in order to get specific information about the project.",
  "As a user, I want to select a pre-defined prompt, in order to get specific information about the project.",
  "As a user, I want the chatbot to provide the source of information, in order to verify the accuracy of the response."
]

# Function to calculate average similarity for a model
def calculate_average_similarity(user_stories):
    scores = []
    for story in user_stories:
        story_embedding = model.encode(story)
        desc_embeddings = model.encode(sentences)
        similarity = util.cos_sim(story_embedding, desc_embeddings).max()
        scores.append(similarity.item())
    return sum(scores) / len(scores)

# Calculate and print average similarity scores for each model
average_gpt = calculate_average_similarity(user_stories_gpt)
average_omni = calculate_average_similarity(user_stories_omni)
average_llama = calculate_average_similarity(user_stories_llama)
average_mixtral = calculate_average_similarity(user_stories_mixtral)

print(f"GPT 3.5 Average Similarity: {average_gpt}")
print(f"Omni Average Similarity: {average_omni}")
print(f"Llama3 70 billion Average Similarity: {average_llama}")
print(f"Mixtral Average Similarity: {average_mixtral}")

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Data setup for all four projects
data = {
    'Project 1': {
        'GPT-3.5': {'Epics': 11, 'User Stories': 11, 'Performance': 5.90, 'Similarity Score': 0.57},
        'GPT-4o': {'Epics': 6, 'User Stories': 18, 'Performance': 16.00, 'Similarity Score': 0.44},
        'Llama': {'Epics': 5, 'User Stories': 12, 'Performance': 3.23, 'Similarity Score': 0.38},
        'Mixtral': {'Epics': 9, 'User Stories': 9, 'Performance': 1.88, 'Similarity Score': 0.36}
    },
    'Project 2': {
        'GPT-3.5': {'Epics': 6, 'User Stories': 7, 'Performance': 5.91, 'Similarity Score': 0.72},
        'GPT-4o': {'Epics': 4, 'User Stories': 9, 'Performance': 11.00, 'Similarity Score': 0.71},
        'Llama': {'Epics': 8, 'User Stories': 8, 'Performance': 3.29, 'Similarity Score': 0.69},
        'Mixtral': {'Epics': 7, 'User Stories': 7, 'Performance': 2.26, 'Similarity Score': 0.72}
    },
    'Project 3': {
        'GPT-3.5': {'Epics': 5, 'User Stories': 5, 'Performance': 10.55, 'Similarity Score': 0.65},
        'GPT-4o': {'Epics': 5, 'User Stories': 8, 'Performance': 16.62, 'Similarity Score': 0.62},
        'Llama': {'Epics': 6, 'User Stories': 6, 'Performance': 3.41, 'Similarity Score': 0.65},
        'Mixtral': {'Epics': 6, 'User Stories': 6, 'Performance': 1.99, 'Similarity Score': 0.55}
    },
    'Project 4': {
        'GPT-3.5': {'Epics': 8, 'User Stories': 8, 'Performance': 9.55, 'Similarity Score': 0.71},
        'GPT-4o': {'Epics': 4, 'User Stories': 8, 'Performance': 16.62, 'Similarity Score': 0.6},
        'Llama': {'Epics': 5, 'User Stories': 7, 'Performance': 3.41, 'Similarity Score': 0.72},
        'Mixtral': {'Epics': 5, 'User Stories': 5, 'Performance': 1.99, 'Similarity Score': 0.55}
    }
}

# Create a 2x2 grid of subplots
fig, axes = plt.subplots(2, 2, figsize=(14, 12), sharex=True, sharey=True)

# Colors for the bar plots
colors = {'Epics': 'blue', 'User Stories': 'green', 'Performance': 'red', 'Similarity Score': 'purple'}

# Loop over projects and plot in a grid
project_keys = list(data.keys())
for i, ax in enumerate(axes.flatten()):
    project = project_keys[i]
    df = pd.DataFrame(data[project]).T  # Transpose to get models as index
    df.plot(kind='bar', ax=ax, color=colors.values(), width=0.8)
    ax.set_title(f'{project}')
    ax.set_ylabel('Values')
    ax.set_xticklabels(df.index, rotation=0)

# General layout adjustments
plt.xlabel('Model')
fig.suptitle('Comparison of Metrics Across Four Projects', fontsize=16)
plt.legend(colors.keys(), loc='upper left', bbox_to_anchor=(1,1))
plt.tight_layout(rect=[0, 0, 1, 0.95])  # Adjust layout to make room for the suptitle
plt.show()
```
